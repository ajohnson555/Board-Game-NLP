```python
import pandas as pd
import numpy as np
```

```python
boardgames = pd.read_csv("cleaned_boardgames.csv")
boardgames
```

```python
group_by_type = boardgames[['type', 'weight', 'rating', 'min_age', 'max_players', 'max_time']].groupby("type").mean().sort_values('weight', ascending=False)
group_by_type.to_csv("grouped_by_type.csv", index=False)
group_by_type
```

```python
type_list = boardgames['type'].unique()
for game_type in type_list:
    games_sorted = boardgames[boardgames['type'] == game_type].sort_values('rating')
games_sorted
```
